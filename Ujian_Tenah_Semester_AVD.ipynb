#**Penanganan Data**

##**Penanganan Missing Value**

Tahap ini merupakan proses Imputasi Data (Missing Value Imputation) sebagai bagian dari pembersihan data. Setelah sebelumnya terdeteksi adanya satu nilai kosong pada kolom Age, kode ini berfungsi untuk mengisi kekosongan tersebut menggunakan nilai Median (nilai tengah) dari distribusi usia responden. Pemilihan median dibandingkan rata-rata (mean) bertujuan untuk menjaga objektivitas data agar tidak terpengaruh oleh nilai ekstrem (outlier), sehingga integritas dataset tetap terjaga dan seluruh baris data (101 baris) dapat digunakan sepenuhnya dalam analisis statistik selanjutnya.

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
```

Insight:

Berdasarkan proses penanganan data yang dilakukan, berikut adalah insight teknisnya:

- Penyelesaian Missing Value: Kolom Age kini telah terisi penuh (101 non-null), sehingga tidak ada lagi data yang "bolong" yang dapat menyebabkan error saat pemrosesan grafik atau pemodelan.

- Stabilitas Distribusi: Penggunaan nilai median memastikan bahwa karakteristik usia responden tetap stabil dan representatif, tanpa menggeser tren demografis utama yang sudah ada sebelumnya.

- Kesiapan Analisis Lanjutan: Dengan selesainya tahap imputasi ini, dataset telah melewati ambang batas kualitas minimum untuk masuk ke tahap transformasi data yang lebih kompleks, seperti standardisasi teks pada nama jurusan atau tahun studi.

- Catatan Teknis: Adanya pesan FutureWarning pada output menunjukkan bahwa metode inplace=True mungkin akan usang di versi Pandas masa depan, namun untuk saat ini, proses pengisian data tetap berhasil dieksekusi dengan sukses pada dataset Anda.

##**Standarisasi Kolom**

Tahap ini merupakan proses Standardisasi dan Harmonisasi Data Kategorikal (Categorical Data Normalization). Berdasarkan temuan pada audit sebelumnya mengenai ketidakkonsistenan penulisan, kode ini berfungsi untuk menyeragamkan format teks menggunakan metode .str.strip() untuk menghapus spasi liar dan .str.lower() untuk menyamakan kapitalisasi. Selain itu, diterapkan teknik Data Mapping menggunakan kamus (course_mapping) untuk menyatukan berbagai variasi nama jurusan yang merujuk pada entitas yang sama (seperti menggabungkan 'engine' dan 'engin' menjadi 'Engineering'). Langkah ini sangat krusial untuk memastikan bahwa saat dilakukan agregasi data, sistem tidak menghitung satu kategori yang sama sebagai beberapa entitas yang berbeda.

```python
df['Your current year of Study'] = df['Your current year of Study'].str.strip().str.lower()
```

##**Relationship/Hubungan**

Tahap ini merupakan proses Analisis Multivariat: Matriks Korelasi (Multivariate Correlation Analysis) menggunakan Heatmap. Karena dataset ini memiliki banyak variabel kategorikal (seperti "Yes/No"), sistem melakukan Label Encoding terlebih dahulu untuk mengubah teks menjadi nilai numerik agar hubungan antar variabel dapat dihitung secara matematis. Fungsi sns.heatmap() kemudian memetakan kekuatan hubungan tersebut ke dalam gradasi warna. Proses ini sangat krusial untuk menemukan "benang merah" atau faktor-faktor yang saling memengaruhi, misalnya apakah status pernikahan memiliki korelasi kuat dengan tingkat depresi atau serangan panik pada mahasiswa.

```python
df_encoded = df.copy()
for col in df_encoded.columns:
    if df_encoded[col].dtype == 'object':
        df_encoded[col] = df_encoded[col].astype('category').cat.codes

# heatmap
plt.figure(figsize=(10,6))
sns.heatmap(df_encoded.corr(), annot=True, cmap='coolwarm')
plt.title('Heatmap Korelasi Antar Variabel')
plt.show()
```

Insight:

Berdasarkan Heatmap Korelasi di atas, berikut adalah temuan hubungan antar variabel yang paling signifikan:

- Korelasi Terkuat (Depresi & Status Pernikahan): Ditemukan korelasi positif yang cukup tinggi (0.6) antara Marital Status dengan Depression. Hal ini memberikan indikasi menarik bahwa status pernikahan memiliki hubungan linear yang lebih kuat terhadap kondisi depresi dibandingkan variabel akademik seperti CGPA.

- Hubungan Antar Gangguan Mental: Variabel Depression, Anxiety, dan Panic Attack menunjukkan korelasi positif yang saling berkaitan (berkisar antara 0.25 hingga 0.27). Artinya, seorang mahasiswa yang mengalami satu jenis gangguan mental memiliki kecenderungan tertentu untuk juga mengalami gangguan mental lainnya.

- Korelasi Akademik yang Lemah: Variabel CGPA dan Age memiliki angka korelasi yang sangat mendekati nol terhadap variabel kesehatan mental. Ini menunjukkan bahwa dalam dataset ini, tingkat kecerdasan akademik atau usia bukan merupakan faktor penentu utama seseorang mengalami depresi atau kecemasan.

- Pencarian Bantuan Spesialis: Terdapat korelasi positif (0.35) antara Marital Status dan Depression terhadap Did you seek any specialist?. Ini mengindikasikan bahwa responden yang sudah menikah atau yang sedang mengalami depresi memiliki kecenderungan sedikit lebih tinggi untuk mencari bantuan medis dibandingkan kelompok lainnya, meskipun secara keseluruhan angkanya masih rendah.
